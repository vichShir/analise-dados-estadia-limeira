### Tratamento textual

```python
df_baratos.loc[1, 'descricao']
```

```python
df_caros.loc[0, 'descricao']
```

```python
# A expressão regular abaixo remove toda palavra com número e torna as palavras minusculas
import re

lower_alpha = lambda x: re.sub(r"""\w*\d\w*""", ' ', x.lower())
df_baratos['descricao'] = df_baratos.descricao.map(lower_alpha)
df_caros['descricao'] = df_caros.descricao.map(lower_alpha)

display(df_baratos.head(2))
display(df_caros.head(2))
```

```python
# Removendo as pontuações
import string

punc_re = lambda x: re.sub('[%s]' % re.escape(string.punctuation), ' ', x)
df_baratos['descricao'] = df_baratos.descricao.map(punc_re)
df_caros['descricao'] = df_caros.descricao.map(punc_re)

display(df_baratos.head(2))
display(df_caros.head(2))
```

```python
df_baratos.loc[1, 'descricao']
```

```python
df_caros.loc[0, 'descricao']
```

```python
# Removendo \n

remove_break_line = lambda x: re.sub(r"""\n""", ' ', x.lower())
df_baratos['descricao'] = df_baratos.descricao.map(remove_break_line)
df_caros['descricao'] = df_caros.descricao.map(remove_break_line)

display(df_baratos.head(2))
display(df_caros.head(2))
```

```python
df_baratos.loc[1, 'descricao']
```

```python
df_caros.loc[0, 'descricao']
```

```python
# Tokenize a revisão em palavras
from nltk.tokenize import word_tokenize

df_baratos['tokens'] = df_baratos.descricao.map(word_tokenize)
df_caros['tokens'] = df_caros.descricao.map(word_tokenize)

display(df_baratos.head(2))
display(df_caros.head(2))
```

```python
# Essa etapa cria uma grande lista combinando os valores dos tokens 
word_list_baratos = sum(df_baratos.tokens.tolist(), [])
word_list_caros = sum(df_caros.tokens.tolist(), [])

print(word_list_baratos[:50])
print(word_list_caros[:50])
```

```python
# Encontrando as palavras mais comuns

from collections import Counter

# Converte a lista em um dicionário com a contagem dos valores
word_counts_baratos = Counter(word_list_baratos)
word_counts_caros = Counter(word_list_caros)

# Inverte chave/valor no dicionário para ordenar
word_counts_baratos = list(zip(word_counts_baratos.values(), word_counts_baratos.keys()))
word_counts_caros = list(zip(word_counts_caros.values(), word_counts_caros.keys()))

# Ordena a lista por contagem
word_counts_baratos = sorted(word_counts_baratos, reverse=True)
word_counts_caros = sorted(word_counts_caros, reverse=True)

# Mostra as 10 palavras mais frequentes
print(word_counts_baratos[:25])
print(word_counts_caros[:25])
```

Acomodações mais baratas, em geral:
  * Não tem ar condicionado
  * Não tem câmeras, interfones ou sistema de monitoramento
  * Não tem faxina inclusa
  * Não tem menção de internet inclusa ou ponto para internet
  * Sem menção de gás
  * Só energia inclusa
  * Lavanderia ou área de serviço coletiva
  * Tem alguma menção de vaga de garagem

## Filtrando acomodações

```python
def search_key_words(df_filtro, key_words):
  search_index = []

  for index in range(0, len(df_filtro)):
    for element in df_filtro.iloc[index].values:
      if type(element) == str:
        if all(x in element.lower() for x in key_words):
          search_index.append(index)

  search_index = set(search_index)
  search_index = sorted(search_index)

  new_df = df_filtro.iloc[search_index, :]
  return new_df
```

```python
df_filtro = search_key_words(df_baratos, ['cozinha', 'banheiro'])
#'cozinha', 'banheiro', 'lavanderia'
#'banheiro', 'cozinha', 'quarto', 'geladeira', 'incluso', 'energia', 'água', 'interfone', 'camera', 'área', 'serviço'
print(df_filtro.shape)
df_filtro
```
